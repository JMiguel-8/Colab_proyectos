#1.Grafica lineal#

Crear una grafica con numeros aleatorios con djoole y colab a partir de un texo y con depuracion de errores en directo

prompt: Buenos dias , me gustaria una grafica lineal con cierto margen de error entre sus puntos y que se visualice como resultado final una linea que recoja la mayor parte de puntos . La visualizacion debe ser tanto grafica como impresa en texto

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
np.random.seed(42)  # Para reproducibilidad
x = np.linspace(0, 10, 50)  # 50 puntos entre 0 y 10

# Ecuación lineal y = 2x + 1 con error aleatorio
y = 2 * x + 1 + np.random.normal(0, 0.5, x.shape)
# Imprimir algunos puntos de datos
print("Algunos puntos de datos (x, y): ")
for i in range(10):  # Imprimir los primeros 10 puntos
    print(f"({x[i]:.4f}, {y[i]:.4f})")
print("Datos generados correctamente.")
```

```python
# Calcular la línea de tendencia
coef = np.polyfit(x, y, 1)
poly1d_fn = np.poly1d(coef)

# Calcular estadísticas
mse = np.mean((y - poly1d_fn(x))**2)
y_mean = np.mean(y)
ss_total = np.sum((y - y_mean)**2)
ss_residual = np.sum((y - poly1d_fn(x))**2)
r_squared = 1 - (ss_residual / ss_total)

print("Cálculos completados.")
```

```python
# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5, label='Datos con ruido')
plt.plot(x, poly1d_fn(x), color='red', label='Línea de tendencia')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal con Datos Aleatorios')
plt.legend()

# Añadir información sobre el modelo en la gráfica
equation = f'y = {coef[0]:.4f}x + {coef[1]:.4f}'
plt.text(0.5, 9, f'Ecuación: {equation}', fontsize=12)
plt.text(0.5, 8, f'Error Cuadrático Medio: {mse:.4f}', fontsize=12)
plt.text(0.5, 7, f'R^2: {r_squared:.4f}', fontsize=12)

plt.ylim(0, 25)  # Ajustar los límites del eje y
plt.show()

# Imprimir algunos puntos de datos
print("Algunos puntos de datos (x, y): ")
for i in range(10):  # Imprimir los primeros 10 puntos
    print(f"({x[i]:.4f}, {y[i]:.4f})")

print(f"\nEcuación de la línea de tendencia: y = {coef[0]:.4f}x + {coef[1]:.4f}")
print(f"\nError cuadrático medio: {mse:.4f}")
print(f"\nCoeficiente de determinación (R^2): {r_squared:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generación de datos
np.random.seed(42)  # Para reproducibilidad
x = np.linspace(0, 100, 50)  # 50 puntos entre 0 y 100

# Función escalar ascendente: y = 0.5x^2 + 2x + 10
y_true = 0.5 * x**2 + 2 * x + 10
y = y_true + np.random.normal(0, 50, x.shape)  # Añadir ruido

# 2. Cálculo de la curva de tendencia
coef = np.polyfit(x, y, 2)  # Ajuste polinómico de grado 2
poly_fn = np.poly1d(coef)

# 3. Cálculo de estadísticas
mse = np.mean((y - poly_fn(x))**2)
y_mean = np.mean(y)
ss_total = np.sum((y - y_mean)**2)
ss_residual = np.sum((y - poly_fn(x))**2)
r_squared = 1 - (ss_residual / ss_total)

# 4. Creación de la gráfica
plt.figure(figsize=(12, 8))
plt.scatter(x, y, color='blue', alpha=0.5, label='Datos con ruido')
plt.plot(x, poly_fn(x), color='red', label='Curva de tendencia')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Polinómica con Datos Aleatorios (Escala Ascendente)')
plt.legend()

# 5. Añadir información sobre el modelo en la gráfica
equation = f'y = {coef[0]:.4f}x^2 + {coef[1]:.4f}x + {coef[2]:.4f}'
plt.text(5, max(y)*0.9, f'Ecuación: {equation}', fontsize=12)
plt.text(5, max(y)*0.85, f'Error Cuadrático Medio: {mse:.4f}', fontsize=12)
plt.text(5, max(y)*0.8, f'R^2: {r_squared:.4f}', fontsize=12)

plt.show()

# 6. Imprimir resultados
print("Algunos puntos de datos (x, y):")
for i in range(10):   # Imprimir los primeros 10 puntos
    print(f"({x[i]:.4f}, {y[i]:.4f})")

print(f"\nEcuación de la curva de tendencia: {equation}")
print(f"Error cuadrático medio: {mse:.4f}")
print(f"Coeficiente de determinación (R^2): {r_squared:.4f}")
```

#4.Creador de graficas curvileas , datos manual , automatico o salir ( tiene un factor estatico para que los puntos tengan poco margen de error a proposito )#

```python
import numpy as np
import matplotlib.pyplot as plt

def get_user_input():
    while True:
        try:
            n = int(input("Ingrese el número de puntos que desea (mínimo 3): "))
            if n < 3:
                print("Por favor, ingrese al menos 3 puntos.")
                continue
            return n
        except ValueError:
            print("Por favor, ingrese un número entero válido.")

def generate_data(n, mode):
    if mode == '1':  # Modo automático con menor margen de error
        x = np.linspace(0, 100, n)
        y = 0.05 * x**2 + 2 * x + np.random.normal(0, 10, n)  # Función más estable con ruido controlado
    else:  # Modo manual
        x = np.linspace(0, 100, n)
        y = []
        for i in range(n):
            while True:
                try:
                    value = float(input(f"Ingrese el valor y para x = {x[i]:.2f}: "))
                    y.append(value)
                    break
                except ValueError:
                    print("Por favor, ingrese un número válido.")
        y = np.array(y)
    return x, y

def plot_regression(x, y):
    coef = np.polyfit(x, y, 2)
    poly_fn = np.poly1d(coef)

    mse = np.mean((y - poly_fn(x))**2)
    r_squared = 1 - (np.sum((y - poly_fn(x))**2) / np.sum((y - np.mean(y))**2))

    plt.figure(figsize=(12, 8))
    plt.scatter(x, y, color='blue', alpha=0.5, label='Datos ingresados')
    plt.plot(x, poly_fn(x), color='red', label='Curva de tendencia')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Regresión Polinómica')
    plt.legend()

    equation = f'y = {coef[0]:.4f}x^2 + {coef[1]:.4f}x + {coef[2]:.4f}'
    plt.text(5, max(y)*0.9, f'Ecuación: {equation}', fontsize=12)
    plt.text(5, max(y)*0.85, f'Error Cuadrático Medio: {mse:.4f}', fontsize=12)
    plt.text(5, max(y)*0.8, f'R^2: {r_squared:.4f}', fontsize=12)
    plt.show()

    print("\nPuntos de datos ingresados (x, y):")
    for i in range(len(x)):
        print(f"({x[i]:.4f}, {y[i]:.4f})")
    print(f"\nEcuación de la curva de tendencia: {equation}")
    print(f"Error cuadrático medio: {mse:.4f}")
    print(f"Coeficiente de determinación (R^2): {r_squared:.4f}")

    print("\nPrograma finalizado.")

# Menú principal
while True:
    print("\nSeleccione el modo de entrada de datos:")
    print("1. Generar datos aleatorios automáticamente")
    print("2. Ingresar datos manualmente")
    print("3. Salir")
    opcion = input("Ingrese su opción: ")

    if opcion == '3':
        print("Saliendo del programa...")
        break
    elif opcion in ['1', '2']:
        n_points = get_user_input()
        x, y = generate_data(n_points, opcion)
        plot_regression(x, y)
        break  # Termina el programa después de mostrar la gráfica
    else:
        print("Opción no válida. Intente nuevamente.")
```
